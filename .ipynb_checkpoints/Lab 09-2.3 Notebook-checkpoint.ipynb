```python
# Import data sets

apple = pd.read_csv('aapl.csv')
apple.columns = ['Date','Open','High','Low','Close','Volume']

google = pd.read_csv('goog.csv')
google.columns = ['Date','Open','High','Low','Close','Volume']

twitter = pd.read_csv('twtr.csv')
twitter.columns = ['Date','Open','High','Low','Close','Volume']

microsoft = pd.read_csv('msft.csv')
microsoft.columns = ['Date','Open','High','Low','Close','Volume']

facebook = pd.read_csv('fb.csv')
facebook.columns = ['Date','Open','High','Low','Close','Volume']

dfs = [apple, google, twitter, microsoft, facebook]
```

```python
def scrubbin(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df['Year'] = df.index.year
    df['Month'] = df.index.month
    df['Day'] = df.index.day
    for i in df:
        if type(df[i][0]) == str:
            df = df[df[i] != '-'].astype(float)
    print df.head()
    print df.info()
    return df

apple = scrubbin(apple)
google = scrubbin(google)
twitter = scrubbin(twitter)
microsoft = scrubbin(microsoft)
facebook = scrubbin(facebook)
```

```python
apple.head()
```

```python
combined = pd.DataFrame()
combined['Open'] = (apple['Open'] + google['Open'] + twitter['Open'] + microsoft['Open'] + facebook['Open']) / 5.0
combined['High'] = (apple['High'] + google['High'] + twitter['High'] + microsoft['High'] + facebook['High']) / 5.0
combined['Low'] = (apple['Low'] + google['Low'] + twitter['Low'] + microsoft['Low'] + facebook['Low']) / 5.0
combined['Close'] = (apple['Close'] + google['Close'] + twitter['Close'] + microsoft['Close'] + facebook['Close']) / 5.0
combined['Volume'] = (apple['Volume'] + google['Volume'] + twitter['Volume'] + microsoft['Volume'] + facebook['Volume']) / 5.0
combined.head()
```

```python
columns = ['Open','High','Low','Close']

def lookin(df):
    for i in columns:
        print i
        df[i].plot()
lookin(microsoft)
plt.show()
```
